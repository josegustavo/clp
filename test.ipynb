```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# %matplotlib inline
# %config InlineBackend.figure_format = 'retina'


def cuboid_data(o, size=(1, 1, 1)):
    l, w, h = size
    x = [[o[0], o[0] + l, o[0] + l, o[0], o[0]],
         [o[0], o[0] + l, o[0] + l, o[0], o[0]],
         [o[0], o[0] + l, o[0] + l, o[0], o[0]],
         [o[0], o[0] + l, o[0] + l, o[0], o[0]]]
    y = [[o[1], o[1], o[1] + w, o[1] + w, o[1]],
         [o[1], o[1], o[1] + w, o[1] + w, o[1]],
         [o[1], o[1], o[1], o[1], o[1]],
         [o[1] + w, o[1] + w, o[1] + w, o[1] + w, o[1] + w]]
    z = [[o[2], o[2], o[2], o[2], o[2]],
         [o[2] + h, o[2] + h, o[2] + h, o[2] + h, o[2] + h],
         [o[2], o[2], o[2] + h, o[2] + h, o[2]],
         [o[2], o[2], o[2] + h, o[2] + h, o[2]]]
    return x, y, np.array(z)


def draw(boxes):
    pallete = sns.color_palette()

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for i in range(len(boxes)):
        box = boxes[i]
        p, s, c = box[0:3], box[3:6], pallete[box[6]]
        X, Y, Z = cuboid_data(p, s)
        ax.plot_surface(X, Y, Z, color=c, alpha=.75,
                        antialiased=0, edgecolor=c)
    ax.set_aspect('equal')
    #ax.view_init(azim=-50, elev=40)
    plt.show()


draw([
    [0, 0, 0, 4, 6, 5, 1], #Primer cubo
    [4, 1, 0, 3, 3, 3.90192, 0], #Segundo cubo
])
```
